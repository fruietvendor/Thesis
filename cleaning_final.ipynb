```python
import os
import pandas as pd
directory_list = ["CE", "SE", "NE_1", "NE_2", "NE_3"]
conversation_id_list = []

for _directory in directory_list:
    file_list = os.listdir(_directory)
    for _file in file_list:
        _filepath = os.path.join(os.path.abspath(_directory), _file)
        fd = pd.read_csv(_filepath)
        fd1 = fd.where(fd.language == "en").dropna(subset = ["language"])
        day_conversation = [conversation for conversation in fd1["conversation_id"].values]
        conversation_id_list.extend(day_conversation)
```

```python
def count_conversations(conversation_list):
    convo_count_dict = {}
    for conversation in conversation_list:
        convo_count_dict.setdefault(conversation,0)
        convo_count_dict[conversation] += 1
        
    convo_count_table = pd.DataFrame(columns = ["conversation_id", "count"])
    convo_count_table["conversation_id"] = list(convo_count_dict.keys())
    convo_count_table["count"] = list(convo_count_dict.values())
    return convo_count_table.set_index("conversation_id")
    
```

```python
import pandas as pd
import os
import nltk
from nltk.corpus import stopwords
from googletrans import Translator
from google.transliteration import transliterate_word
from textblob import TextBlob
tweet_list = []

sw = stopwords.words("english")

class strip_dataset:
    
    
    def __init__(self, data_directory, filename, necessary_columns = ["tweet"]):
        
        self.filename = filename
        self.dirname = data_directory
        self.filepath = os.path.join(os.path.abspath(self.dirname), self.filename)
        self.necessary_columns = necessary_columns
        self.dataframe = pd.read_csv(self.filepath)
        self.cleaned_df = self.data_cleaner()
        self.csv = self.push_to_csv()
        self.to_list = self.push_to_list()

    
    @staticmethod
    def tweet_cleaner(tweet):
        tweet = tweet.lower()
        tweet = tweet.replace(",", " ")
        tweet = tweet.replace("'", "")
        tweet = tweet.replace(".", " ")
        splits = tweet.split()
        splits = [split for split in splits if split.isalpha() and len(split) > 1 and split[0] != "@" \
                  and split[0] != "#" and split not in sw]
        return splits
    
        
        
    def data_cleaner(self):
        _cleaned = self.dataframe.copy()
        _cleaned = _cleaned.where(_cleaned.language == "en").dropna(subset = ["language"])
        _cleaned = _cleaned[self.necessary_columns]
        _cleaned["cleaned_tweet"] = _cleaned["tweet"].apply(self.tweet_cleaner)
        _cleaned = _cleaned.reset_index(drop = True)
        return _cleaned
    

        
    def push_to_csv(self):    
        self.cleaned_df.to_csv("cleaned_tweets.csv", mode = 'a', index = False, header = False)
        with open("cleaned_tweets.csv") as f:
            f.close()
            
      
    def push_to_list(self):
        for row in self.cleaned_df["cleaned_tweet"]:
            tweet_list.append(row)
        
```

```python
def streaming_data(directory_list):
    for _directory in directory_list:
        file_list = os.listdir(_directory)
        for _file in file_list:
            strip_dataset(_directory, _file)
            
```

```python
sse = []
for k in range(1,11):
    kmeans = KMeans(
        init = "k-means++",
        n_init = 10,
        n_clusters = k,
        max_iter = 300,
        random_state = 42)
    kmeans.fit(word_vectors.vectors)
    sse.append(kmeans.inertia_)
    
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_coeffs = []
for k in range(2, 21):
    kmeans = KMeans(
        init  = "k-means++",
        n_init = 10,
        max_iter = 300,
        n_clusters = k,
        random_state = 42)
    kmeans.fit(word_vectors.vectors)
    score = silhouette_score(word_vectors.vectors, kmeans.labels_)
    silhouette_coeffs.append(score)
```

```python
plt.style.use("fivethirtyeight")
plt.plot(range(2, 21), silhouette_coeffs)
plt.xticks(range(2, 21))
plt.xlabel("no. of clusters")
plt.ylabel("silhouette coeffs")
plt.show()
```
